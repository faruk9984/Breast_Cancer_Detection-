# Split DatFrame in Train and Test

```python
X = cancer_df.drop(['target'], axis = 1) 
X.head(6)
```

```python
y = cancer_df['target'] 
y.head(6)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 51)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

# Machine Learning Model Building

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

# Suppor vector Classifier

```python
from sklearn.svm import SVC
svc  = SVC()
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
svc2 = SVC()
svc2.fit(X_train_sc, y_train)
y_pred_sc = svc2.predict(X_test_sc)
accuracy_score(y_test, y_pred_sc)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(random_state = 51, penalty = 'l2')
lr.fit(X_train, y_train)
y_pred_lr= lr.predict(X_test)
accuracy_score(y_test, y_pred_lr)
```

```python
lr2 = LogisticRegression(random_state = 51, penalty = 'l2')
lr2.fit(X_train_sc, y_train)
y_pred_lr_sc = lr2.predict(X_test_sc)
accuracy_score(y_test, y_pred_lr_sc)
```

# K – Nearest Neighbor Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
accuracy_score(y_test, y_pred_knn)
```

```python
knn2 = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
knn2.fit(X_train_sc, y_train)
y_pred_knn_sc = knn2.predict(X_test_sc)
accuracy_score(y_test, y_pred_knn_sc)
```

```python
knn3 = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
knn3.fit(X_train_sc, y_train)
knn3.score(X_test_sc,y_test)
```

# Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb  = GaussianNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)
accuracy_score(y_test, y_pred_nb)
```

```python
nb2 = GaussianNB()
nb2.fit(X_train_sc, y_train)
y_pred_nb_sc = nb2.predict(X_test_sc)
accuracy_score(y_test, y_pred_nb_sc)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt  = DecisionTreeClassifier(criterion = 'entropy', random_state = 51)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
accuracy_score(y_test, y_pred_dt)
```

```python
dt2 = DecisionTreeClassifier(criterion = 'entropy', random_state = 51)
dt2.fit(X_train_sc, y_train)
y_pred_dt_sc = dt2.predict(X_test_sc)
accuracy_score(y_test, y_pred_dt_sc)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf  = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 51)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
accuracy_score(y_test, y_pred_rf)
```

```python
rf2 = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 51)
rf2.fit(X_train_sc, y_train)
y_pred_rf_sc = rf2.predict(X_test_sc)
accuracy_score(y_test, y_pred_rf_sc)
```
